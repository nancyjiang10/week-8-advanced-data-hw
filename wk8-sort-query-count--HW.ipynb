````python
## import winter data
## import winter data
dfw = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-winter.csv")
dfw
````

## Get a sense of the data

````python
## SUMMER
## what exactly do we have: columns, datatypes, etc.
dfs.info()
````

````python
## Create a dataframe that for only summer games years and cities
subset_of_full_data_df = dfw[["Year", "City", "Sport"]]
subset_of_full_data_df
````

````python
dfs.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze' " ).value_counts().to_frame("Quantity").sort_values(by="Quantity", ascending = True)
````

````python
dfs.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze' " ).value_counts(ascending = False).to_frame("Quantity")
````

````python
dfs.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze' " ).value_counts(ascending = True)
````

Homework Challenge

2. Top 10 count of gold medals by country
3. Of these top 10, what percent of the gold medals do they have?

````python
## Medals handed out in summer games
dfs["Medal"].value_counts(ascending = False).to_frame("Quantity")
````

````python
## Medals handed out in winter games
dfw["Medal"].value_counts(ascending = False).to_frame("Quantity")
````

````python
dfs["Season"] = "summer"
dfs.head()
````

````python
## In the winter dataframe, add "Season" column with the value "Winter"
dfw["Season"] = "winter" 
dfw.head()
````

````python
dfw.columns 
````

````python
dfw.columns ==  dfs.columns
````

````python
## join winter and summer dataframes together into a dataframe called df
# df = pd.concat([dfs,dfw]) ## Led to the wrong total index. 
df = pd.concat([dfs,dfw], ignore_index = True)
df
````

````python
summer_df = df.query("Season == 'summer'")
summer_df
````

````python
summer_df.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze' " ).sort_values(by="Medal")
````

````python
##Summer medals count in absolute
summer_df["Medal"].value_counts(ascending = True)
````

````python
##Summer medals in percentage
summer_df["Medal"].value_counts(normalize = True)
````

````python
winter_df = df.query("Season == 'winter'")
winter_df
````

````python
##Winter medals count in absolute
winter_df["Medal"].value_counts(ascending = True)
````

````python
##Winter medals in percentage
winter_df["Medal"].value_counts(normalize = True)
````

````python
##All gold medals subset
gold_df = df.query("Medal == 'Gold'")
gold_df
````

````python
##Top gold medals by country
top_gold_countries = gold_df["NOC"]
top_gold_countries
````

````python
top_gold_countries.head(10)
````

````python
## return as percentage, rounded 2 decimal place
````

````python
round(top_gold_countries.head(10).value_counts(normalize = True) * 100, 2)
````
